Основные понятия теории вероятности
===================================

1. Выпишите 25 первых цифр десятичного представления числа $\pi$.
    1. Вы выбираете наугад цифру из этого набора. Какова вероятность получить каждую из 10 цифр.
    1. Какая наиболее вероятная цифра? Медианная? Средняя?
    1. Найдите стандартное отклонение для этого распределения.

1. Стрелка сломанного спидометра может свободно двигаться, отскакивая от обоих ограничителей. Таким образом если ударить по стрелке, то она остановится на любом угле от 0 до $\pi$.
    1. Каково распределение вероятности $\rho(\theta)$? Убедитесь, что полная вероятность равна единице.
    1. Вычислите $\langle\theta \rangle$, $\langle\theta^2 \rangle$ и $\sigma$.
    1. Вычислите $\langle sin \theta \rangle$, $\langle cos \theta \rangle$, и $\langle cos^2\theta \rangle$.

1. Рассмотрим тот же сломанный спидометр из прошлой задачи теперь интересуясь $x$-координатой стрелки.
    1. Каково распределение вероятности $\rho(x)$? Нарисуйте его от $-2r$ до $2r$.
    1. Вычислите $\langle x \rangle$, $\langle x^2 \rangle$ и $\sigma$. Как эти величины связаны с аналогичными из прошлой задачи?

1. Рассмотрим гауссово распределение $\rho(x) = A e^{- \lambda (x-a)^2}.$
    1. Нормируйте это распределение на единицу.
    1. Вычислите $\langle x \rangle$, $\langle x^2 \rangle$ и $\sigma$.
    1. Нарисуйте график.
    1. Найдите $\sigma$ в случае если $\rho(x)$ — распределение вероятности нахождения электрона в основном состоянии атома водорода.

1. Игла Бюффона. Иголка длины $l$ бросается случайным образом на лист бумаги с начерченными параллельными линиями с расстоянием $l$ между линиями. Какова вероятность иголке пересечь линию?
